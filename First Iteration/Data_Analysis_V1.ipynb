```python
# Import the pandas library for data handling
import pandas as pd
```

```python
# Load a CSV file named 'livreDA.csv' into a pandas DataFrame
# The 'sep' parameter specifies that the columns are separated by a semicolon (';')
data = pd.read_csv("livreDA.csv" , sep = ";") 
# Display the contents of the DataFrame
data
```

```python
# Generate descriptive statistics of the DataFrame 'data'
data.describe()
```

```python
# Retrieve the column names of the DataFrame 'data'
data.columns
```

```python
# Check for missing values in the DataFrame 'data'
# This will return the count of null values for each column
data.isnull().sum()
```

```python
# Check for duplicate rows in the DataFrame 'data'
data.duplicated().sum()
```

```python
# Create a list of specific book IDs to filter the DataFrame
bookid = [1068, 38338, 19100]
# Filter the DataFrame 'data' to include only rows where 'bookID' is in the 'bookid' list
data[data['bookID'].isin(bookid)]
```

```python
# Remove the rows with indices 327, 5288, and 9610 from the DataFrame 'data'
# The 'inplace=True' argument modifies the original DataFrame without creating a copy
data.drop([327, 5288, 9610], inplace = True)
```

```python
data
```

```python
# Convert the 'isbn' column in the DataFrame 'data' to string type
data['isbn'] = data['isbn'].astype(str)
# Apply a lambda function to modify the 'isbn' values:
# Remove the last character if it ends with 'X'
data['isbn'] = data['isbn'].apply(lambda x: x[:-1] if x.endswith('X') else x)
# Display the first 5 rows of the modified DataFrame
data.head(5)
```

```python
# Removal of 2 rows (publisher): oct-18
# Create a list of specific book IDs to filter the DataFrame
bookid = [26012, 28225]
# Filter the DataFrame 'data' to include only rows where 'bookID' is in the 'bookid' list
data[data['bookID'].isin(bookid)]
```

```python
# Remove the rows with indices 6903 and 7328 from the DataFrame 'data'
data.drop([6903, 7328], inplace = True)
```

```python
data
```

```python
# Drop the 'isbn13' column from the DataFrame 'data'
data = data.drop(columns=['isbn13'])
```

```python
data
```

```python
data.isnull().sum()
```

```python
# Remove rows with missing values from the DataFrame 'data'
data.dropna()
```

```python
# Import the matplotlib library for data visualization
# The 'pyplot' module is typically used for creating plots and charts
import matplotlib.pyplot as plt
```

```python
##Number of Books per Language
books_per_language = data['language_code'].value_counts()

# Display the results
books_per_language.plot(kind='bar', title='Number of Books per Language')
plt.xlabel('Language Code')
plt.ylabel('Number of Books')
plt.show()
```

```python
# Number of books per publisher
books_per_publisher = data['publisher'].value_counts().head(10)  # Show the 10 most frequent publishers

# Show the results
books_per_publisher.plot(kind='bar', title='Number of books per publisher')
plt.xlabel('Publisher')
plt.ylabel('numbers of books')
plt.show()
```

```python
# Extract the publication year
data['publication_year'] = pd.to_datetime(data['publication_date'], errors='coerce').dt.year

# Number of books published per year
books_per_year = data['publication_year'].value_counts().sort_index()

# Show the results
books_per_year.plot(kind='line', title='Number of Books Published per Year')
plt.xlabel('Year')
plt.ylabel('Number of Books')
plt.show()
```

```python
# Number of books per author
books_per_author = data['authors'].value_counts().head(10)  # Display the 10 most prolific authors

# Display the results
books_per_author.plot(kind='bar', title='Nombre de livres par auteur')
plt.xlabel('Auteur')
plt.ylabel('Nombre de livres')
plt.show()
```
