**Genre Analysis**

Let's start by analysing the genre over time to continue upon earlirt analysis

```python
def plot_genre_ratings_over_time(genre_column, genre_name):
    genre_data = data[data[genre_column] == 1]
    ratings_over_time = genre_data.groupby('publication_year')['average_rating'].mean()

    plt.plot(ratings_over_time.index, ratings_over_time.values, marker='o', label=genre_name)


# Plotting ratings over time for selected genres
plt.figure(figsize=(14, 8))

# Fantasy/Paranormal
plot_genre_ratings_over_time('is_fantasy_paranormal', 'Fantasy/Paranormal')

# Mystery/Thriller/Crime
plot_genre_ratings_over_time('is_mystery_thriller_crime', 'Mystery/Thriller/Crime')

# Romance
plot_genre_ratings_over_time('is_romance', 'Romance')

# Children
plot_genre_ratings_over_time('is_children', 'Children')
plt.title('Average Ratings Over Time by genre')
plt.xlabel('Publication Year')
plt.ylabel('Average Rating')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting ratings over time for selected genres
plt.figure(figsize=(14, 8))

# Comic
plot_genre_ratings_over_time('is_comic', 'Comic')

# History Biography
plot_genre_ratings_over_time('is_history_biography', 'History Biography')

# Poetry
plot_genre_ratings_over_time('is_poetry', 'Poetry')

# Young Adult
plot_genre_ratings_over_time('is_young_adult', 'Young Adult')

plt.title('Average Ratings Over Time by genre')
plt.xlabel('Publication Year')
plt.ylabel('Average Rating')
plt.legend()
plt.grid(True)
plt.show()
```

Overall genres seem to follow the same movement, we can see that history biography books between 1900 and 1960 have higher average ratings but Young adult books ten to have better rating afrer 2000

* Let's now take a look At the distribution of average ratings by genre and see what we can take from it

```python
from pypalettes import get_hex

palette = get_hex("Alacena")
genre_columns = [
    'is_children', 'is_comic', 'is_fantasy_paranormal',
    'is_history_biography', 'is_mystery_thriller_crime',
    'is_poetry', 'is_romance', 'is_young_adult'
]
# Melting the dataset to have a 'Genre' column for easier plotting
melted_genres = data.melt(id_vars=['average_rating'], value_vars=genre_columns,
                              var_name='genre', value_name='is_genre')

# Filter only rows where the book belongs to a genre (is_genre == 1)
melted_genres = melted_genres[melted_genres['is_genre'] == 1]

# Rename genre columns for better readability in the plot
genre_labels = {
    'is_children': 'Children',
    'is_comic': 'Comic',
    'is_fantasy_paranormal': 'Fantasy/Paranormal',
    'is_history_biography': 'History/Biography',
    'is_mystery_thriller_crime': 'Mystery/Thriller/Crime',
    'is_poetry': 'Poetry',
    'is_romance': 'Romance',
    'is_young_adult': 'Young Adult'
}

melted_genres['genre'] = melted_genres['genre'].map(genre_labels)

# Plotting the distribution of average ratings by genre
plt.figure(figsize=(14, 8))
sns.boxplot(data=melted_genres, x='genre', y='average_rating', palette=palette)
plt.title('Distribution of Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

The boxplot shows the distribution of average ratings across different genres.

The most important observations we can makes are!

*   Fantasy/Paranormal and Children's genres show a relatively wide distribution of ratings, indicating varied reader reception.

*   Poetry has a narrower distribution, with a higher median rating and fewer outliers we can say it is genrally liked by the readers.

*   Mystery/Thriller/Crime and History/Biography have a more concentrated distribution, though with some lower outliers.

**Potential Analysis and Reasons:**

####Holiday Season (October to December):
- We can obseve a spike in average ratings towards the end of the year, where readers might be more generous with ratings, or publishers might release higher-quality books anticipating the holiday demand.

####Post-Holiday Slump (January to February):
- The lower ratings at the beginning of the year could be a result of readers being more critical after the holiday season, or fewer high-profile books being released during these months.
####Publishing Strategies:

- End-of-Year Releases: Publishers often release their most anticipated books in the last quarter of the year to capitalize on holiday sales
- Summer Releases: There is a slight increase in ratings around the mid-year, which could be linked to the summer reading season, where lighter or more enjoyable reads might be released.

###Themes:
- Books released in specific months might align with seasonal themes, such as cozy reads in the winter months, that appeal more to readers and result in higher ratings.

*Might be insteresting to add a season column to the dataset*

```python

# First, let's add a column for the publication month from the dataset
# Assuming there's a 'publication_month' column or similar

# Convert publication month to seasons
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

data['season'] = data['publication_month'].apply(month_to_season)

# Calculate average rating by season
avg_rating_by_season = data.groupby('season')['average_rating'].mean()

# Plotting average rating by season
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20, 6))
sns.barplot(x=avg_rating_by_season.index, y=avg_rating_by_season.values, palette=palette)
plt.title('Average Rating by Season')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()

```

```python

def month_to_trimester(month):
    if month in [1, 2, 3, 4]:
        return '1st'
    elif month in [5, 6, 7,8]:
        return '2nd'
    elif month in [9, 10, 11, 12]:
        return '3rd'

data['trimester'] = data['publication_month'].apply(month_to_trimester)

# Calculate average rating by trimester
avg_rating_by_trimester = data.groupby('trimester')['average_rating'].mean()


plt.figure(figsize=(20, 6))
sns.barplot(x=avg_rating_by_trimester.index, y=avg_rating_by_trimester.values, palette=palette)
plt.title('Average Rating by Trimester')
plt.xlabel('Trimester')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

The variation can not be seen here maybe this is due to the fact that the average ratings are skewed to the right and are mostly high in this dataset

```python
# Group data by publication year and month, then calculate the average rating
avg_rating_per_month = data.groupby(['publication_year', 'publication_month'])['average_rating'].mean().reset_index()

# Create a new column for 30-year intervals
avg_rating_per_month['year_interval'] = (avg_rating_per_month['publication_year'] // 30) * 30

# Group by year interval and month, then calculate the average rating
avg_rating_per_month_interval = avg_rating_per_month.groupby(['year_interval', 'publication_month'])['average_rating'].mean().reset_index()

# Plotting the average rating per month for every 30 years
plt.figure(figsize=(15, 8))
sns.lineplot(data=avg_rating_per_month_interval, x='publication_month', y='average_rating', hue='year_interval', marker='o')
plt.title('Average Rating per Month for Every 30 Years')
plt.xlabel('Publication Month')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

This let us think that the second interval 1920-1950 might be the on who caused the average rating per month seen earlier as we can see it has an overall increasing trend with a lot of fluctuations

The Hypotheses of earlier about season also seem to be not valid as over the years average rating seems to be nearly consistant over months espescially for the last three intervals (1950-1980, 1980-2010, 2010).

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load your dataset
books_df = data.copy()

# Add a feature representing the length of the book title
books_df['title_length'] = books_df['title'].apply(len)

# Select relevant features
features = [
    'title_length', 'text_reviews_count', 'num_pages', 'publication_day',
    'publication_month', 'publication_year', 'ratings_count',
    'authors_average_rating', 'authors_text_reviews_count',
    'authors_ratings_count', 'country_code', 'language_code', 'is_ebook',
     'is_children', 'is_comic',
    'is_fantasy_paranormal', 'is_history_biography', 'is_mystery_thriller_crime',
    'is_poetry', 'is_romance', 'is_young_adult'
]

# Simplify by using Label Encoding for categorical features
for col in ['country_code', 'language_code', ]:
    le = LabelEncoder()
    books_df[col] = le.fit_transform(books_df[col].astype(str))

# Select features and target
X = books_df[features]
y = books_df['average_rating']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Extract and display feature importances
feature_importances = model.feature_importances_
feature_importances_df = pd.DataFrame({'feature': X.columns, 'importance': feature_importances})

# Sort by importance
feature_importances_df = feature_importances_df.sort_values(by='importance', ascending=False)

# Display the feature importances
print(feature_importances_df)
```
